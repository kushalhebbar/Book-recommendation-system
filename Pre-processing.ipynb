```python
import pandas as pd

#Reading original data files
books = pd.read_csv('books.csv', sep=';', error_bad_lines=False, encoding="latin-1")
users = pd.read_csv('users.csv', sep=';', error_bad_lines=False, encoding="latin-1")
ratings = pd.read_csv('ratings.csv', sep=';', error_bad_lines=False, encoding="latin-1")

books.columns = ['Book', 'bookTitle', 'bookAuthor', 'yop', 'publisher', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L']
users.columns = ['User', 'Location', 'Age']
ratings.columns = ['User', 'Book', 'Rating']

books.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L'],axis=1,inplace=True)
```

```python
import numpy as np

#Removing years from yop column that are outliers and retaining only years till 2012
books.loc[(books.yop > 2012) | (books.yop == 0), 'yop'] = np.NAN
books.yop.fillna(round(books.yop.mean()), inplace=True)
books.yop = books.yop.astype(np.int32)
```

```python
users.loc[(users.Age < 8) | (users.Age > 90), 'Age'] = np.nan 
users.Age = users.Age.fillna(users.Age.mean())
users.Age = users.Age.astype(np.int32)
```

```python
#Find intersection between users dataset and books dataset

#books.ISBN and users.userID

updated_ratings = ratings[ratings.Book.isin(books.Book)]
updated_ratings = updated_ratings[updated_ratings.User.isin(users.User)]
```

```python
sparsity = 1.0 - len(updated_ratings) / float(users.shape[0]  * books.shape[0])
print(str(sparsity * 100) + '%')
```

```python
explicit_ratings = updated_ratings[updated_ratings.Rating != 0]
explicit_ratings.head(20)
```

```python
#merge books and explicit_ratings based on ISBN

final = pd.merge(books, explicit_ratings, on = "Book")
final.drop(['bookTitle', 'bookAuthor', 'yop', 'publisher', 'Book'],axis=1,inplace=True)
final.columns = ['Book', 'User', 'Rating']
final.head()
```

```python
books.to_csv('new_books.csv', index=False)
final.to_csv('final.csv', index=False)
```
